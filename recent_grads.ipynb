## Making use of Jupyter magic line

Because of the Jupyter magic line '%matplotlib inline', the scatter plot is displayed immediately. This means that we can write one line of code to generate a scatter plot, run the cell using a keyboard shortcut, inspect the plot, and repeat.

```python
recent_grads.plot(x='Sample_size', y='Employed',
                 kind='scatter', title='Employed vs. Sample_size',
                 figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size', y='Median',
                 kind='scatter', title='Median v Sample_size',
                 figsize=(5,10))
```

```python
recent_grads.plot(x='Full_time', y='Median',
        kind='scatter', title='Median v Full_time',
        figsize=(5,10))
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate',
                 kind='scatter',
                 title='Unemployment_rate v ShareWomen',
                 figsize=(5,10))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter',
                 title='Median v Men', figsize=(5,10))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter',
                 title='Median v Women', figsize=(5,10))
```
